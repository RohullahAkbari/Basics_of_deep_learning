```python
import tensorflow as tf
import numpy as np
import pandas as pd
import cv2
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder , LabelEncoder

tf.compat.v1.disable_eager_execution()
```

```python
room_path = os.listdir('./datasets/living_room/')
room_path
```

```python
room = []

for item in room_path:

    all_img = os.listdir('./datasets/living_room' + '/' + item)

    for img in all_img:
        room.append((item, str('./datasets/living_room' + '/' + item + '/' + img)))
        print(room[:1])
```

```python
room_df = pd.DataFrame(data=room, columns=['room type', 'image'])
print(room_df.head())
print(room_df.tail())
```

```python
print('total number of images in dataset: ', len(room_df))
```

```python
room_count = room_df['room type'].value_counts()

print('room is each category:')
print(room_count)
```

```python
path = 'datasets/living_room/'

image = []
label = []

for i in room_path:
    data_path = path + str(i)
    filename = [i for i in os.listdir(data_path)]
    # print(i)
    # print(filename)
    
    for f in filename:
        img = cv2.imread(data_path + '/' + f)
        img = cv2.resize(img, (32, 32))
        image.append(img)
        label.append(i)
```

```python
image = np.array(image)
image.shape
```

```python
image = image.astype('float32') / 255.0
```
