```python
import tensorflow as tf
import keras
import numpy as np
import pandas as pd
import cv2
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder , LabelEncoder
from sklearn.utils import shuffle
from PIL import Image


tf.compat.v1.disable_eager_execution()
```

```python
room_path = os.listdir('./datasets/living_room/')
room_path
```

```python
room = []

for item in room_path:

    all_img = os.listdir('./datasets/living_room' + '/' + item)

    for img in all_img:
        room.append((item, str('./datasets/living_room' + '/' + item + '/' + img)))
        print(room[:1])
```

```python
room_df = pd.DataFrame(data=room, columns=['room type', 'image'])
print(room_df.head())
print(room_df.tail())
```

```python
print('total number of images in dataset: ', len(room_df))
```

```python
room_count = room_df['room type'].value_counts()

print('room is each category:')
print(room_count)
```

```python
path = 'datasets/living_room/'

image = []
label = []

for i in room_path:
    data_path = path + str(i)
    filename = [i for i in os.listdir(data_path)]
    # print(i)
    # print(filename)
    
    for f in filename:
        img = cv2.imread(data_path + '/' + f)
        img = cv2.resize(img, (32, 32))
        image.append(img)
        label.append(i)
```

```python
image = np.array(image)
image.shape
```

```python
image = image.astype('float32') / 255.0
```

```python
image.shape
```

```python
y = room_df['room type']
y[:5]
```

```python
y_labelEncoder = LabelEncoder()
y = y_labelEncoder.fit_transform(y)

y
```

```python
image , y = shuffle(image, y, random_state=42)

X_train , X_test, y_train, y_test = train_test_split(image, y , test_size=.2, random_state=425)
```

```python
X_train.shape , y_train.shape , X_test.shape , y_test.shape
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape = (32, 32, 3)),

    keras.layers.Dense(256, activation=tf.nn.tanh),

    keras.layers.Dense(2, activation=tf.nn.softmax)
])
```

```python
model.summary()
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=5)
```

```python
y_pred = model.predict(X_test)
y_pred
```
