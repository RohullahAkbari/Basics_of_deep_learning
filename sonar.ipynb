```python
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
```

```python
sonar_df = pd.read_csv('sonar.csv', header=None)
sonar_df.head()
```

```python
X = sonar_df[sonar_df.columns[:60]].values

y = sonar_df[sonar_df.columns[60]]
```

```python
X.shape
```

```python
y.shape
```

### To change the labels in to numeric values

```python
y_labelEncoder = LabelEncoder()
y = y_labelEncoder.fit_transform(y)

y
```

```python
y.shape
```

```python
y = y.reshape(-1,1)

ct = ColumnTransformer(
    transformers=[('encoder', OneHotEncoder(), [0])],
    remainder='passthrough'
)

y = ct.fit_transform(y)

y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42, shuffle=True)
```

```python
X_train.shape, X_test.shape, y_test.shape, y_train.shape
```

```python
lr = .01
epochs = 100
cost_history = np.empty(shape=[1], dtype=float)
```

```python
n_class = 2

tf.compat.v1.disable_eager_execution()

x = tf.compat.v1.placeholder(tf.float32, [None, 60])

w = tf.Variable(tf.cast(tf.zeros([60, n_class]), tf.float32))

b = tf.Variable(tf.cast(tf.zeros([n_class]), tf.float32))

# initialize all variables

init = tf.compat.v1.global_variables_initializer()

y = tf.compat.v1.placeholder(tf.float32, [None, n_class])

pred = tf.nn.softmax(tf.matmul(x, w) + b)
```

```python
# fine the cost function

cost_function = tf.reduce_mean(tf.reduce_sum(tf.math.log(pred)))
```

```python
# for minimize the cost function we should use the optimizers like (GradientDescentopitmizer)

train_steps = tf.compat.v1.train.GradientDescentOptimizer(lr).minimize(cost_function)
```

```python
# start the sessions

sess = tf.compat.v1.Session()
sess.run(init)
mse_history = []
```

```python
for i in range(epochs):
    
    sess.run(train_steps, feed_dict={x: X_train, y: y_train})

    cost = sess.run(cost_function, feed_dict={x: X_test, y: y_test})
    cost_history = np.append(cost_history, cost)
    

    y_pred = sess.run(pred, feed_dict={x: X_test})
    print('epoch: ', i, 'cost: ', cost)

    mse = tf.reduce_mean(tf.square(y_pred - y_test))
    mse_history.append(sess.run(mse))
```
